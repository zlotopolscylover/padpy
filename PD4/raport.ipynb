## Testowanie jakości każdego algorytmu na zbiorach benchmarkowych

### knn dla k=5

```python
n = ["SAheart","BreastCancer","ecoli","Glass","Ionosphere","iris",
     "pimaDiabetes_norm","seeds","sonar-all-data","spam","wdbc",
     "Wine","wine_1","wine_2","wine_3","wine_4","wine_5"]
dict = []
for index in n:
    X = []
    with open('{}.data.csv'.format(index), 'r') as infile:
        for line in infile:
            X.append(list(map(float, line.strip().split(","))))
        
    y = []
    with open('{}.labels.csv'.format(index), 'r') as infile:
        for line in infile:
            y.append(int(line.strip()))

    dict.append(CV.crossvalidation(X, y, 5))
pd.DataFrame(data=dict) 
```
